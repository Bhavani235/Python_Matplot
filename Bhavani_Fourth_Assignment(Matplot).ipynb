```python
# Import Required Libraries
import matplotlib.pyplot as plt
import numpy as np # This line imports the numpy library and assigns it to the alias 'np'

x = np.linspace(0, 10, 100)
y = np.sin(x)  # Sine of x

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)

# Add title and labels
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Bar Plot

# Create data for the bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='green')

# Add title and labels
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# Histogram

# Create data for the histogram
data = np.random.randn(100)  # 1000 random values from a normal distribution

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', edgecolor='black')

# Add title and labels
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Pie Chart

# Create data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.3, 0, 0, 0)  # explode the 1st slice (i.e. 'Category A')

# Create the pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Add title
plt.title('Pie Chart')

# Show the plot
plt.show()
```

```python
# Box Plot

# Create data for the box plot
np.random.seed(10)
data_box = np.random.randn(100)

# Create the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_box)

# Add title and labels
plt.title('Box Plot')
plt.xlabel('X-axis')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# Heatmap

# Create data for the heatmap
np.random.seed(0)
data_heatmap = np.random.rand(10, 12)

# Create the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(data_heatmap, cmap='viridis', aspect='auto')

# Add color bar
plt.colorbar()

# Add title and labels
plt.title('Heatmap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
# 3D Plot

from mpl_toolkits.mplot3d import Axes3D

# Create data for the 3D plot
x_3d = np.linspace(-5, 5, 100)
y_3d = np.linspace(-5, 5, 100)
x_3d, y_3d = np.meshgrid(x_3d, y_3d)
z_3d = np.sin(np.sqrt(x_3d**2 + y_3d**2))

# Create the 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_3d, y_3d, z_3d, cmap='viridis')

# Add title and labels
ax.set_title('3D Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```

```python

# Bubble Plot
plt.figure()
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000  # Bubble sizes
colors = np.random.rand(50)  # Bubble colors

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()  # Show color scale
plt.title('Bubble Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python


# Donut Chart
plt.figure()
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Donut Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
